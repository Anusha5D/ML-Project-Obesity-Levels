```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("C://Users/anush/Desktop/ObesityDataSet_raw_and_data_sinthetic.csv")

df.head()
```

```python
# Check for Missing Values
print(df.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Example categorical columns to encode
categorical_columns = ['Gender', 'family_history_with_overweight']

# Apply Label Encoding
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

```

```python
# Apply One-Hot Encoding
import pandas as pd

# Print DataFrame before One-Hot Encoding
print("DataFrame before One-Hot Encoding:")
print(df.head())

# Specify categorical columns for One-Hot Encoding
categorical_columns = ['FAVC', 'FCVC', 'CAEC', 'SMOKE', 'CH2O', 'SCC', 'FAF', 'TUE', 'CALC', 'MTRANS']

# Apply One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

# Print DataFrame after One-Hot Encoding
print("\nDataFrame after One-Hot Encoding:")
print(df_encoded.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

print("DataFrame before Normalization:")
print(df[['Age', 'Height', 'Weight']].describe())

df[['Age', 'Height', 'Weight']] = scaler.fit_transform(df[['Age', 'Height', 'Weight']])

print("DataFrame after Normalization:")
print(df[['Age', 'Height', 'Weight']].describe())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

print("DataFrame before Standardization:")
print(df[['Age', 'Height', 'Weight']].describe())

df[['Age', 'Height', 'Weight']] = scaler.fit_transform(df[['Age', 'Height', 'Weight']])

print("DataFrame after Standardization:")
print(df[['Age', 'Height', 'Weight']].describe())
```

```python
import numpy as np

Q1 = df[['Age', 'Height', 'Weight']].quantile(0.25)
Q3 = df[['Age', 'Height', 'Weight']].quantile(0.75)
IQR = Q3 - Q1

print("Outliers before handling:")
outliers = ((df[['Age', 'Height', 'Weight']] < (Q1 - 1.5 * IQR)) | (df[['Age', 'Height', 'Weight']] > (Q3 + 1.5 * IQR)))
print(outliers.sum())

# Handle outliers (e.g., remove them)
df = df[~((df[['Age', 'Height', 'Weight']] < (Q1 - 1.5 * IQR)) | (df[['Age', 'Height', 'Weight']] > (Q3 + 1.5 * IQR))).any(axis=1)]

print("DataFrame after handling outliers:")
print(df[['Age', 'Height', 'Weight']].describe())
```
