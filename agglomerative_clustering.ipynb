```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r'/Users/hibaali/Desktop/Coding.html/python/ObesityDataSet_raw_and_data_sinthetic.csv')
```

```python
# Extract Age and Weight columns
age_weight_data = data[['Age', 'Weight']].values

# Standardize the features
scaler = StandardScaler()
age_weight_scaled = scaler.fit_transform(age_weight_data)
```

```python
# Apply Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3)  # Assuming 3 clusters
cluster_labels_aw = agg_clustering.fit_predict(age_weight_scaled)
```

```python
# Calculate silhouette score
silhouette_avg_aw = silhouette_score(age_weight_scaled, cluster_labels_aw)
```

```python
# Plotting function for clusters
def plot_clusters(X, labels=None, title=''):
    sns.set(style='whitegrid')
    plt.figure(figsize=(10, 6))
    
    if labels is None:
        plt.scatter(X[:, 0], X[:, 1], c='grey', s=50, alpha=0.7)
        plt.title(f'{title} (Without Clustering)')
    else:
        sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, palette='deep', s=60, alpha=0.7)
        plt.title(f'{title} (With Clustering)')
        plt.legend(title='Cluster')
    
    plt.xlabel('Age (standardized)')
    plt.ylabel('Weight (standardized)')
    plt.show()
```

```python
# Visualizing the data without clustering
plot_clusters(age_weight_scaled, title='Age vs. Weight Dataset')
```

```python
# Visualizing the data with clustering
plot_clusters(age_weight_scaled, cluster_labels_aw, title='Age vs. Weight Dataset')
```

```python
# Display the silhouette score
print(f'Silhouette Score for Age and Weight Clustering: {silhouette_avg_aw:.3f}')
```
